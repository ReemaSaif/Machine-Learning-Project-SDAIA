```python
import gradio as gr
import joblib
import pandas as pd

model = joblib.load("randomforest_model.pkl")


def predict_rentals(season, year, month, hour, holiday, weekday, workingday,
                    weather_situation, temperature, feels_like_temp,
                    humidity, windspeed):

    input_data = pd.DataFrame([{
        "season": season,
        "year": year,
        "month": month,
        "hour": hour,
        "holiday": holiday,
        "weekday": weekday,
        "workingday": workingday,
        "weather_situation": weather_situation,
        "temperature": temperature,
        "feels_like_temp": feels_like_temp,
        "humidity": humidity,
        "windspeed": windspeed
    }])

    input_data["is_weekend"] = input_data["weekday"].isin([0, 6]).astype(int)

    input_data["is_rush_hour"] = input_data["hour"].isin([7, 8, 9, 16, 17, 18, 19]).astype(int)

    input_data["season_peak"] = (input_data["season"] == 3).astype(int)

    input_data["temperature_bin"] = pd.cut(input_data["temperature"],bins=[0, 0.3, 0.6, 1],labels=["cold", "mild", "hot"],ordered=True)


    prediction = model.predict(input_data)[0]
    return int(prediction)
```

```python
# Gradio UI

inputs = [
    gr.Dropdown([1, 2, 3, 4], label="Season"),
    gr.Dropdown([0, 1], label="Year (0 = 2011, 1 = 2012)"),
    gr.Dropdown(list(range(1, 13)), label="Month"),
    gr.Dropdown(list(range(0, 24)), label="Hour"),
    gr.Dropdown([0, 1], label="Holiday"),
    gr.Dropdown(list(range(0, 7)), label="Weekday"),
    gr.Dropdown([0, 1], label="Working Day"),
    gr.Dropdown([1, 2, 3, 4], label="Weather Situation"),
    gr.Slider(0, 1, step=0.01, label="Temperature"),
    gr.Slider(0, 1, step=0.01, label="Feels-like Temperature"),
    gr.Slider(0, 1, step=0.01, label="Humidity"),
    gr.Slider(0, 1, step=0.01, label="Windspeed")
]

output = gr.Number(label="Predicted Bike Rentals")

app = gr.Interface(
    fn=predict_rentals,
    inputs=inputs,
    outputs=output,
    title="Bike Rental Prediction (Random Forest)",
    description="Predict hourly bike rental demand using weather and time features."
)

app.launch(debug=True)
```
